```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Baca gambar
image = cv2.imread(r'D:\Documents\AMIKOM\Semester 6\PROYEK DATA MINING\klasifikasi foto\ipynb\GAMBAR\0_IPHONE-SE_S.JPG')

# Konversi gambar ke skala abu-abu
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Tampilkan gambar
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
```

```python
# Terapkan filter Gaussian
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Tampilkan gambar setelah filter Gaussian
plt.imshow(blurred_image, cmap='gray')
plt.title('Gambar setelah Filter Gaussian')
plt.axis('off')
plt.show()
```

```python
# Terapkan deteksi tepi Canny
canny_edges = cv2.Canny(blurred_image, 30, 150)

# Tampilkan hasil deteksi tepi Canny
plt.imshow(canny_edges, cmap='gray')
plt.title('Deteksi Tepi Menggunakan Canny')
plt.axis('off')
plt.show()
```
